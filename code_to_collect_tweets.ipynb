```python
import os
import pandas as pd
from datetime import date
```

```python

end_date = '2022-09-31'
```

```python
search_term = 'rāsəst' or 'racism' or 'Racism' or 'racist'
from_date = '2020-01-01'
```

```python
#total number of tweets for search term

os.system(f"snscrape --since {from_date} twitter-search '{search_term} until:{end_date}' > result-tweets.csv")
if os.stat("result-tweets.csv").st_size == 0:
    counter = 0
else:
    df = pd.read_csv('result-tweets.csv',names=['link'])
    counter = df.size
print('Number of Tweets : '+ str(counter))
```

```python
max_results =200000
```

```python
extracted_tweets_5 = "snscrape --format '{content!r}'"+ f" --max-results {max_results} --since {from_date} twitter-search '{search_term} until:{end_date}' > extracted-tweets_5.csv"
os.system(extracted_tweets_5)
if os.stat("extracted-tweets_5.csv").st_size == 0:
  print('No Tweets found')
else:
  df = pd.read_csv('extracted-tweets_5.csv', names=['content'])
  for row in df['content'].iteritems():
    print(row)
```
